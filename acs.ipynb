```python
import pandas as pd
```

```python
df = pd.read_csv('data/raw/acs_16_ntx_counties.csv') #read in 2016 data

cols = df.columns #create list of column names

cols = [col.lower().replace(' ', '_') for col in cols] #convert columns to snake case
cols = [col.lower().replace('(', '').replace(')', '') for col in cols] #strip parentheses

for i,j in enumerate(cols): #remove Census prefixes 
    if '!!' in j:
        cols[i] = j.split('!!')[-1]
    else:
        pass

df.columns = cols #rename columns to cleaned values
df['label_grouping'] = df['label_grouping'].str.replace(' -> ', '')
county = '' #create temp variable to hold county name
for i,j in enumerate(df['label_grouping']):
    if ' County, Texas' in j: #data is not structured correctly, so we need to clean it up
        county = j.split(' ')[0] #grab just the county name
    else:
        df['label_grouping'][i] = county.strip() + '_' + j.strip() #add county name to the front of the label

df['label_grouping'] = df['label_grouping'].str.replace(' County, Texas', '') #clean up label grouping column
df['label_grouping'] = df['label_grouping'].str.replace('_  ', '_')

df = df.dropna(subset=['total_population']).reset_index(drop=True) #drop nan rows
df = df.dropna(axis=1).reset_index(drop=True) #drop nan columns
df['year'] = 2016 #add year column

df.to_csv('data/cleaned/acs_16_ntx_counties_cleaned.csv', index=False) #save cleaned data
df
```

```python
df1 = pd.read_csv('data/raw/acs_21_ntx_counties.csv')
cols = df1.columns

cols = [col.lower().replace(' ', '_') for col in cols]
cols = [col.lower().replace('(', '').replace(')', '') for col in cols]
for i,j in enumerate(cols):
    if '!!' in j:
        cols[i] = j.split('!!')[-1]
    else:
        pass
df1.columns = cols
county = ''
for i,j in enumerate(df1['label_grouping']):
    if ' County, Texas' in j:
        county = j.split(' ')[0]
    else:
        df1['label_grouping'][i] = county.strip() + '_' + j.strip()

df1['label_grouping'] = df1['label_grouping'].str.replace(' County, Texas', '')
df1['label_grouping'] = df1['label_grouping'].str.replace('_  ', '_')

df1 = df1.dropna(subset=['total_population']).reset_index(drop=True)
df1 = df1.dropna(axis=1).reset_index(drop=True)
df1['year'] = 2021

df1.to_csv('data/cleaned/acs_21_ntx_counties_cleaned.csv', index=False)
df1
```

```python
df = df.loc[:,~df.columns.duplicated()].copy() #remove duplicate columns
df1 = df1.loc[:,~df1.columns.duplicated()].copy()
df = pd.concat([df, df1]).reset_index(drop=True) #concatenate dataframes
```

```python
for i in df.columns: #remove commas and percent signs from string columns
    if type(df[i][0]) == str: 
        df[i] = df[i].str.replace(',', '')
        df[i] = df[i].str.replace('%', '')
    else:
        pass
cols = list(df.columns)
cols = [i for i in cols if i != 'label_grouping']
cols
for i in cols: #convert number columns to floats
    df[i] = pd.to_numeric(df[i], errors='coerce')
df
```

```python
df = df.loc[df['label_grouping'].str.contains('Estimate')].reset_index(drop=True)
df = df[['label_grouping', 'year', 'total_population', 'male', 'female', 'median_age_years', '18_years_and_over', 'white', 'black_or_african_american', 'asian', 'two_or_more_races', 'hispanic_or_latino_of_any_race', 'total_housing_units', 'american_indian_and_alaska_native']]
df['children'] = df['total_population'] - df['18_years_and_over']
df = df.sort_values(by=['label_grouping', 'year']).reset_index(drop=True)
df
```

## Percent Change

```python
ds = [] #create list to store dictionaries
cols = list(df.columns) #create list of column names

for i in list(df['label_grouping'].unique()): #loop through each county
    d = {} #create dictionary
    d['county'] = i #add county name to dictionary
    for j in cols: #loop through each numeric column
        if j not in ['label_grouping', 'year']: #skip non-numeric columns
            try: #calculate percent change
                d[j] =  ((float(df.loc[((df['label_grouping'] == i) & (df['year'] == 2021)),  j])) - float(df.loc[((df['label_grouping'] == i) & (df['year'] == 2016)),  j]))  / float(df.loc[((df['label_grouping'] == i) & (df['year'] == 2016)),  j]) * 100
            except ZeroDivisionError: #if one of the values is zero, set percent change to zero
                d[j] = 0
        else:
            pass
    
    ds.append(d) #add dictionary to list
dfx = pd.DataFrame(ds) #create dataframe from list of dictionaries
cols = ['change_' + i for i in cols] #add change_ prefix to column names
cols = ['county'] + cols #rename columns
cols.remove('change_label_grouping') #remove label_grouping column
cols.remove('change_year') #remove year column
dfx.columns = cols
dfx
dfx.to_csv('data/created/acs_16_21_percent_change.csv', index=False) #save dataframe to csv
```
